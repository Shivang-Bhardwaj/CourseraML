```python
df = pd.read_csv(r"https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                 names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"])
df.head()
```

## Basic Scatter plots with Matplotlib

```python
plt.plot(df.sepal_length,
        df.sepal_width,
        ls='', marker = "o")
```

## Scatter Plots with Multiple Layers

```python
plt.plot(df.sepal_length,
        df.sepal_width,
        ls='', marker = "o",
        label = "sepal")
plt.plot(df.petal_length,
        df.petal_width,
        ls='', marker = "o",
        label = "petal")
plt.legend()
```

## Histograms

```python
plt.hist(df.sepal_length, bins=25)
plt.show()
```

## Customizing Plots

```python
# matplotlib syntax
fig, ax = plt.subplots()
ax.barh(np.arange(10),
       df.sepal_width.iloc[:10])

# set position of ticks ans tick label
ax.set_yticks(np.arange(0.4, 10.4, 1.0))
ax.set_yticklabels(np.arange(1,11))
ax.set(xlabel = 'xlabel', ylabel = 'ylabel', title = 'Title')

plt.show()
```

## Customizing Plots: by Group

```python
df2 = df.groupby("species").mean()
df2
```

```python
df2.plot(color = ['red','blue','black','green'],
        fontsize = 10.0, figsize = (4,4))
```

## Pair Plots for features

```python
# seaborn plot, feature correlations
sns.pairplot(df, hue='species', height=3)
```

## Hexabin Plot

```python
sns.jointplot(x=df.sepal_length,
             y = df.sepal_width, kind = 'hex')
```

darker the hexagon, there will be more values(denser) 

## Facet Grid 

```python
plot = sns.FacetGrid(df,
                    col='species',
                    margin_titles=True)
plot.map(plt.hist, 'sepal_length', color='green')

plot = sns.FacetGrid(df,
                    col='species',
                    margin_titles=True)
plot.map(plt.hist, 'sepal_width', color='blue')
```
